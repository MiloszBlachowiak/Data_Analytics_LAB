2. Using numpy.random generate arrays

```python
interval = [0, 1]
A = np.random.uniform(low=interval[0], high=interval[1], size=1000)
```

```python
mean = 0
std_dev = 1
B = np.random.normal(loc=mean, scale=std_dev, size=1000)
```

```python
mean = 0
std_dev = 10
C = np.random.normal(loc=mean, scale=std_dev, size=1000)
```

```python
alpha = [1.0, 2.0, 3.0]
beta = [1.0, 2.0, 9.0]
D1 = np.random.beta(a=alpha[0], b=beta[0], size=1000)
D2 = np.random.beta(a=alpha[1], b=beta[1], size=1000)
D3 = np.random.beta(a=alpha[2], b=beta[2], size=1000)
```

```python
mean = [0, 0]
cov1 = [[1, 0], [0, 1]]
cov2 = [[2, 1], [1, 2]]

E1 = np.random.multivariate_normal(mean=mean, cov=cov1, size=10000)
E2 = np.random.multivariate_normal(mean=mean, cov=cov2, size=10000)
```

3. For one dimensional distributions plot histograms. Assign bin values to represent the geometry of distribution. Compare the histograms (normalized) with plots of analytical formulas for distributions (see eg. Wikipedia).

```python
# uniform distribution
n_bins=25
plt.figure()
A_hist, A_bins, p = plt.hist(A, bins=n_bins)
plt.title("Uniform distribution histogram")
plt.xlabel("Value")
plt.ylabel("Number of samples")
plt.show()
```

```python
# normal distribution (mean=0, std_dev = 1)
n_bins=25
plt.figure()
B_hist, B_bins, p = plt.hist(B, bins=n_bins)
plt.title("Normal distribution histogram")
plt.xlabel("Value")
plt.ylabel("Number of samples")
plt.show()
```

```python
# normal distribution (mean=0, std_dev = 10)
n_bins=25
plt.figure()
C_hist, C_bins, p = plt.hist(C, bins=n_bins)
plt.title("Normal distribution histogram")
plt.xlabel("Value")
plt.ylabel("Number of samples")
plt.show()
```

```python
# beta distribution
n_bins=25
plt.figure()
D1_hist, D1_bins, p = plt.hist(D1, bins=n_bins)
plt.title("Beta distribution histogram nr 1")
plt.xlabel("Value")
plt.ylabel("Number of samples")
plt.show()

# beta distribution
n_bins=25
plt.figure()
D2_hist, D2_bins, p = plt.hist(D2, bins=n_bins)
plt.title("Beta distribution histogram nr 2")
plt.xlabel("Value")
plt.ylabel("Number of samples")
plt.show()

# beta distribution
n_bins=25
plt.figure()
D3_hist, D3_bins, p = plt.hist(D3, bins=n_bins)
plt.title("Beta distribution histogram nr 3")
plt.xlabel("Value")
plt.ylabel("Number of samples")
plt.show()
```
